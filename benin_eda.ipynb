```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
plt.style.use("seaborn")
```

```python
sns.set()
```

```python
df = pd.read_csv("data/benin_raw.csv", parse_dates=["Timestamp"])
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
missing = df.isna().sum().sort_values(ascending=False)
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({"missing": missing, "missing_pct": missing_pct})
missing_df
```

```python
high_nulls = missing_df[missing_df["missing_pct"] > 5]
high_nulls
```

```python
key_cols = ["GHI", "DNI", "DHI", "ModA", "ModB", "WS", "WSgust"]
```

```python
from scipy import stats

z_scores = df[key_cols].apply(lambda x: stats.zscore(x, nan_policy='omit'))
outlier_mask = (z_scores.abs() > 3)
```

```python
outlier_counts = outlier_mask.sum()
outlier_counts
```

```python
df_clean = df[~outlier_mask.any(axis=1)].copy()
df_clean.shape
```

```python
df_clean = df_clean.drop(columns=["Comments"])
```

```python
df_clean.to_csv("data/benin_clean.csv", index=False)
```

```python
df_clean.describe()
```

```python
df_clean["Timestamp"] = pd.to_datetime(df_clean["Timestamp"])
df_clean = df_clean.set_index("Timestamp")
```

```python
df_clean[["GHI", "DNI", "DHI", "Tamb"]].resample("H").mean().plot(figsize=(12,6))
plt.title("Hourly Solar Irradiance - Benin")
plt.show()
```

```python
df_clean.groupby("Cleaning")[["ModA", "ModB"]].mean()
```

```python
sns.boxplot(data=df_clean, x="Cleaning", y="ModA")
plt.show()

sns.boxplot(data=df_clean, x="Cleaning", y="ModB")
plt.show()
```

```python
corr_cols = ["GHI", "DNI", "DHI", "TModA", "TModB", "Tamb"]
sns.heatmap(df_clean[corr_cols].corr(), annot=True)
plt.show()
```

```python
sns.scatterplot(data=df_clean, x="WS", y="GHI")
plt.show()

sns.scatterplot(data=df_clean, x="RH", y="GHI")
plt.show()
```

```python
df_clean["GHI"].hist(bins=40)
plt.show()

df_clean["WS"].hist(bins=40)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df_clean["Tamb"], df_clean["GHI"], s=df_clean["RH"], alpha=0.3)
plt.xlabel("Temperature (Tamb)")
plt.ylabel("GHI")
plt.title("Bubble Chart: GHI vs Tamb sized by RH")
plt.show()
```
